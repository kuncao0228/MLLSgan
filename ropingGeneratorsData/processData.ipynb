```python
import cv2
import os
import numpy as np
import shutil 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
DATA_DIR = 'toys_data'
GRAY_DIR = 'gray_data'
OUTPUT_DIR = '../Train_AB/train'
FULL_DIR = 'toys_data_small'
```

```python
for i in (os.listdir(DATA_DIR)):

    if i.endswith('.png'):
        img = cv2.imread(os.path.join(DATA_DIR, i))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        gray[0:25, 0:25] = 0
        gray[50:75, 0:25] = 0
        gray[100:125, 0:25] = 0
        gray[150:175, 0:25] = 0
        gray[200:225, 0:25] = 0


        gray[0:25, 50:75] = 0
        gray[50:75, 50:75] = 0
        gray[100:125, 50:75] = 0
        gray[150:175, 50:75] = 0
        gray[200:225, 50:75] = 0


        gray[0:25, 100:125] = 0
        gray[50:75, 100:125] = 0
        gray[100:125, 100:125] = 0
        gray[150:175, 100:125] = 0
        gray[200:225, 100:125] = 0


        gray[0:25, 150:175] = 0
        gray[50:75, 150:175] = 0
        gray[100:125, 150:175] = 0
        gray[150:175, 150:175] = 0
        gray[200:225, 150:175] = 0

        gray[0:25, 200:224] = 0
        gray[50:75, 200:224] = 0
        gray[100:125, 200:224] = 0
        gray[150:175, 200:224] = 0
        gray[200:225, 200:224] = 0

        blur = cv2.blur(gray,(7, 7))
        cv2.imwrite(os.path.join(GRAY_DIR, i), blur)
        
        
```

```python
for i in (os.listdir(DATA_DIR)):

    if i.endswith('.png'):
        image1 = cv2.imread(os.path.join(DATA_DIR, i))
        image2 = cv2.imread(os.path.join(GRAY_DIR, i))
        concat = np.concatenate([image1, image2], 1)
        cv2.imwrite(os.path.join(OUTPUT_DIR, i), concat)
        
```

```python
#####CREATE TEST SET

TEST_DIR = 'test_data'
GRAY_TEST_DIR = 'gray_test_data'
FINAL_TEST_DIR = '../Train_AB/test'
```

```python
for i in (os.listdir(FULL_DIR)):
    if "." not in i:
        classPath = os.path.join(FULL_DIR, i)
        for j in os.listdir(classPath):
            if "." not in j and int(j) == 1:
                exposurePath = os.path.join(classPath, j)
                count = 0
                for element in os.listdir(exposurePath):
                    if count == 4:
                        break
                    else:

                        currentPath = os.path.join(exposurePath, element)
                        destinationPath = os.path.join(TEST_DIR, str(i) + "_" + str(element))
#                         print(destinationPath)
#                         print(currentPath)
                        
                        shutil.copyfile(currentPath, destinationPath) 
                        count+=1
```
