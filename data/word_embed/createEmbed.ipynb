```python
import csv
```

```python
with open('images.txt') as f:
    lines = [line.strip().split()[1].split("/")[1] for line in f.readlines()]
```

```python
lines
```

```python
with open('img_attr.csv') as f:
    attributes = [line.strip().split(",")[1] for line in f.readlines()][1:]
```

```python
attributes
```

```python
dict_att = {}
with open('attributes.txt') as f:
    for line in f.readlines():
        dict_att[line.strip().split()[0]] = line.strip().split()[1].split("::")[1]
```

```python
dict_att
```

```python
attributes = [dict_att[att] for att in attributes]
```

```python
dict_embed = {}
with open('prim_colors.vec') as f:
    for line in f.readlines():
        dict_embed[line.strip().split()[0]] = list(map(float, line.strip().split()[1:]))
embeds = [dict_embed[att] for att in attributes] 
```

```python
embed_data = dict(zip(lines,embeds))
```

```python
import json

# as requested in comment

with open('embed.txt', 'w') as file:
     file.write(json.dumps(embed_data))
```
